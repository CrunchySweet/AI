```python
import pandas as pd
import seaborn as sns
```

```python
edf = pd.read_json('english_movies.json')
fdf = pd.read_json('french_movies.json')
```

```python
edf.info(), fdf.info()
```

```python
sns.regplot(data= edf, x="budget", y="revenue"), sns.regplot(data= fdf, x="budget", y="revenue")
print("English Correlation:", edf["budget"].corr(edf['revenue']),
      "French Correlation:", fdf["budget"].corr(fdf['revenue']))
```
